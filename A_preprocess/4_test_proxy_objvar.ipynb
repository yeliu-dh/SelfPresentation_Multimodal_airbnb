```python
proxy_vars=['price',"availability_90"]
## price:
df['price'] = (
    df['price']
    .replace('[\$,]', '', regex=True)  # 去掉 $ 和 ,
    .astype(float)                     # 转成数值
)


print(f"len avant de filtrer 'price', 'availability_90': {len(df)}")
df=df[(df["price"]!=None) and df['availability_90']!=None]
print(f"apres: {len(df)}")

for var in proxy_vars:
    print(df[var].dtype)
    if df[var].dtype =="int64" or df[var].dtype =="float64":#!="object"
        print(df[var].value_counts(dropna=False),"\n")
        print(df[var].describe(include='all'),"\n")

    else :             
        print(df[var].value_counts(dropna=False),"\n")
    print("-----------------------------------------------------------")
```
