```python
df=pd.read_csv("data_jo\listings_paris2406.csv")
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')

start_date='2024-01-01'
end_date='2024-06-30'
#7.26 JO 
newhost_jo = df[df['host_since'].between(start_date, end_date)]

# 也可以画一个时间序列直方图
import matplotlib.pyplot as plt

newhost_jo['host_since'].hist(bins=60)
plt.title('Host Since Distribution (Paris)')
plt.xlabel('Year')
plt.ylabel('Number of hosts')
plt.xticks(rotation=45)
plt.show()
```

```python
# paris 2409
df=pd.read_csv("data_jo\listings_paris2409.csv")
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')

newhost=df[df['host_since'].between('2024-01-01', '2024-08-31')]

# 按月汇总
monthly_counts = (
    newhost['host_since']
    .dt.to_period('M')     # 转为年月周期（如 2024-03）
    .value_counts()
    .sort_index()          # 排序按时间
)

# 绘图
monthly_counts.plot(kind='bar', figsize=(10,5))
plt.title('Monthly New Hosts (Paris 2024)')
plt.xlabel('Month')
plt.ylabel('Number of New Hosts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
